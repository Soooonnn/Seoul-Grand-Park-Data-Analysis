## 계절성과 추세 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

data = pd.read_csv('merged_data.csv')
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)

def plot_trend_seasonal(series, freq=12):
    decomposition = seasonal_decompose(series.dropna(), model='additive', period=freq)

    plt.figure(figsize=(12, 6))

    plt.subplot(2, 1, 1)
    plt.plot(decomposition.trend, label='Trend', color='blue')
    plt.title('Trend')
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.plot(decomposition.seasonal, label='Seasonal', color='green')
    plt.title('Seasonal')
    plt.legend()

    plt.tight_layout()
    plt.show()

for column in data.columns:
    print(f"==== {column} ====")

    plt.figure(figsize=(10, 6))
    plt.plot(data[column].dropna(), label='Original')
    plt.title(f'Time Series Plot of {column}')
    plt.legend()
    plt.show()

    plot_trend_seasonal(data[column], freq=12)
```

```python
from statsmodels.tsa.stattools import adfuller, kpss

def adf_kpss_test(series):
    adf_result = adfuller(series.dropna())
    adf_stat = adf_result[0]
    adf_p_value = adf_result[1]
    adf_crit_values = adf_result[4]

    print(f"ADF Statistic: {adf_stat}")
    print(f"p-value: {adf_p_value}")
    print("Critical Values:")
    for key, value in adf_crit_values.items():
        print(f"   {key}: {value}")

    kpss_result = kpss(series.dropna(), regression='c', nlags="auto")
    kpss_stat = kpss_result[0]
    kpss_p_value = kpss_result[1]
    kpss_crit_values = kpss_result[3]

    print(f"\nKPSS Statistic: {kpss_stat}")
    print(f"p-value: {kpss_p_value}")
    print("Critical Values:")
    for key, value in kpss_crit_values.items():
        print(f"   {key}: {value}")
    
    print("="*40)

for column in data.columns:
    print(f"==== {column} ====")
    adf_kpss_test(data[column])
```

### PM2.5

```python
order = (1, 1, 2) 
model = ARIMA(data['pm2.5'], order=order)

model_fit = model.fit()

forecast_steps = 7
pred_pm25 = model_fit.forecast(steps=forecast_steps)

forecast_series_pm25 = pd.Series(pred_pm25)

forecast_dates = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=forecast_steps, freq='D')

plt.figure(figsize=(10, 6))
plt.plot(data.index, data['pm2.5'], label='Actual', color='blue')
plt.plot(forecast_dates, forecast_series_pm25, label='Forecast', color='red', linestyle='--')
plt.title('ARIMA Forecast of PM2.5 for Next 7 Days')
plt.xlabel('Date')
plt.ylabel('PM2.5 Value')

plt.legend()

plt.show()

print(forecast_series_pm25)
X_test['pm2.5'] = forecast_series_pm25
```

## SARIMAX로 예측한 피쳐들을 가지고 RandomForest + Grid Search를 사용하여 타겟값(입장객 수) 예측

```python
target_data = pd.read_excel('20년도 1월 1일~7일 입장객 수.xlsx')
target_data['date'] = pd.to_datetime(target_data['date'])
target_data.set_index('date', inplace=True)
```

```python
target_data
```

```python
from sklearn.preprocessing import RobustScaler
X = data.iloc[:,:-1]
y = data.iloc[:,-1]

scaler = RobustScaler()
scaler.fit(X)

X_scaled = scaler.transform(X)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rf = RandomForestRegressor()
param_grid = {
    'n_estimators':[100,150,200],
    'max_depth':[None, 10,20,30],
    'min_samples_split':[2,5,10,20],
    'min_samples_leaf':[1,2,4,10]
}

grid_search = GridSearchCV(estimator=rf,
                          param_grid = param_grid,
                          scoring = 'neg_mean_squared_error',
                          cv = 5,
                          verbose=2,
                          n_jobs=-1)
grid_search.fit(X_scaled, y)
print('Best parameters: ', grid_search.best_params_)

best_rf = grid_search.best_estimator_

y_pred_model = best_rf.predict(X_test_scaled)
```

```python
target_data['SARIMAX'] = forecast_series_y
```

```python
target_data['Random Forest + Grid Search'] = y_pred_model
```

## 실제 입장객수 데이터와 SARIMAX모델을 사용하여 예측한 입장객수, SARIMAX로 예측한 Feature들을 가지고 RandomForest + Grid Search를 사용하여 예측한 데이터 간의 성능 비교

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model_performance(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)  # RMSE는 MSE의 제곱근
    r2 = r2_score(y_true, y_pred)

    print(f'MAE: {mae}')
    print(f'MSE: {mse}')
    print(f'RMSE: {rmse}')
    print(f'R²: {r2}')

print("SARIMAX: ")
evaluate_model_performance(target_data['actual'], target_data['SARIMAX'])
print("\n")
print("Random Forest + Grid Search: ")
evaluate_model_performance(target_data['actual'], target_data['Random Forest + Grid Search'])
```
